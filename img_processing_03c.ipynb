# original image

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline

img = mpimg.imread('lizzards.jpg')    
gray = np.dot(img, [0.299, 0.587, 0.144])   
plt.imshow(gray, cmap = plt.get_cmap('gray'))
plt.show()
print gray.shape
```

# Laplacian hakatan

```python
import scipy.ndimage as ndimage
def laplacian(x):
    return max(x[1] + x[3] - 4*x[4] + x[5] + x[7], 0)
edges = ndimage.generic_filter(gray, laplacian, size=(3,3))
edgy = np.zeros(img.shape)
edgy [...,0] = edges
plt.imshow(edges,  cmap = plt.get_cmap('gray'))
plt.show()
```
