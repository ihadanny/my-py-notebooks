# original image + histogram

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.144])
img = mpimg.imread('150110-173028.jpg')     
gray = rgb2gray(img)    
plt.imshow(gray, cmap = plt.get_cmap('gray'))
plt.show()
plt.hist(gray)
plt.show()
```

# histogram equalization

```python
h = np.histogram(gray, bins=256, range=(0.0, 255.0))
c = 255.0*np.cumsum(h[0])/sum(h[0])
plt.plot(h[1][0:256], c)
plt.show()
```

# show equalized image + histogram

```python
new_img = np.zeros(gray.shape)
for j in range(gray.shape[1]):
    for i in range(gray.shape[0]):
        new_img[i][j] = c[gray[i][j]]
plt.imshow(new_img, cmap = plt.get_cmap('gray'))
plt.show()
plt.hist(new_img)
plt.show()
```
