```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import random
import math
import seaborn as sns

sns.set_style("whitegrid")
sns.set_style("ticks")
```

```python
data = []
th = 4.0
outliers_prob = 0.03
X = np.array([])
Y = np.array([])
for i in xrange(100):
    x = random.uniform(-10.0, 10.0)
    if (x > th and random.uniform(0.0, 1.0) < outliers_prob) or (x <= th and random.uniform(0.0, 1.0) > outliers_prob):
        y = 1.0
    else:
        y = 0.0        
    X=np.append(X, x)
    Y=np.append(Y, y)
```

```python
plt.scatter(X, Y)
plt.grid()
plt.show()
```

```python
w = [-1.0, -1.0]
res = 100
```

```python
X_PR = np.linspace(-10.0, 10.0, res)
Y_PR = np.array([1.0/(1+math.exp(-w[0]-w[1]*x)) for x in X_PR ])
#Y_PR = np.array([w[0]+w[1]*x for x in X_PR ])
plt.plot(X_PR, Y_PR)
plt.gca().set_ylim([-1,2])
plt.scatter(X, Y)
plt.grid()
plt.show()
```

```python
W0 = np.linspace(-10.0, 10.0, res)
SE,LL,GRAD = [],[],[]
for w0 in W0:
    Y_PR = np.array([1.0/(1+math.exp(-w0-w[1]*x)) for x in X ])
    #Y_PR = np.array([w0+w[1]*x for x in X_PR ])    
    SE.append(sum((Y_PR - Y)**2))
    LL.append(sum(-Y*np.log(Y_PR) - (1.0 - Y)*np.log(1.0-Y_PR)))
    GRAD.append()
    
#plt.scatter(W0, SE, color="red")
plt.scatter(W0, LL, color="blue")
plt.grid()
plt.show()    
    
```
