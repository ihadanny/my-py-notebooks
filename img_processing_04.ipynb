# read the image into 8x8 tiles

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import matplotlib.image as mpimg
%matplotlib inline

im = mpimg.imread("lizzards.jpg")
gray = np.dot(im, [0.299, 0.587, 0.144])  
plt.imshow(gray, cmap = plt.get_cmap('gray'))

height, width = gray.shape
eight = 80
tiles = {}

for y in xrange(height):
    for x in xrange(width):
        current_tile = (x/eight, y/eight)
        if not current_tile in tiles:
            tiles[current_tile] = np.zeros((eight,eight))
        tiles[current_tile][(y%eight)][x%eight] = gray[y][x]            
```

# dct -> quantize -> idct

```python
from scipy.fftpack import dct, idct
import heapq
dct_tiles = {}
new_tiles = {}

def visualize_dct(d):
    d = np.log(abs(d).clip(0.1))
    maxi, mini = d.max(), d.min()
    d = 255*(d - mini)/(maxi-mini)
    return d

for key, tile in tiles.iteritems():
    d = dct(dct(tile,axis=0), axis=1)
    for y in xrange(eight):
        for x in xrange(eight):
            if x > 30 or y > 30: d[y][x] = 0
    new_tiles[key] = idct(idct(d, axis=1), axis=0)/(eight*eight)
    dct_tiles[key] = visualize_dct(d)

plt.imshow(dct_tiles[(3,4)], cmap = plt.get_cmap('gray'))
plt.show()
```

# convert back to image

```python
pixels_wow = np.zeros(gray.shape)
pixels_d = np.zeros(gray.shape)
for y in xrange(height):
    for x in xrange(width):
        current_tile = (x/eight, y/eight)
        pixels_d[y][x] = int(dct_tiles[current_tile][y%eight][x%eight])
        p = int(new_tiles[current_tile][y%eight][x%eight])
        if x%eight <= 1 or y%eight <= 1: 
            p = 0
        pixels_wow[y][x]=p

plt.imshow(pixels_wow, cmap = plt.get_cmap('gray'))

```
