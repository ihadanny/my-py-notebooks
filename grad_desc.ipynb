# Generating some data with a single feature X and a binary response Y


X is continuous and sampled uniformly from -10 to 10

Y is binary distirbuted with a clear relation to X. 

$y = 1|x < th$ 

$y = 0|x >= th$ 

and some noise. note that data is generated IID.

```python
th = 4.0
noise_prob = 0.05
X = np.array([])
Y = np.array([])
for i in xrange(100):
    x = random.uniform(-10.0, 10.0)
    if x <= th:
        y = 1.0
    else:
        y = 0.0
    if random.uniform(0.0, 1.0) < noise_prob:
        y = 1.0 - y
    X=np.append(X, x)
    Y=np.append(Y, y)
```

```python
plt.scatter(X, Y)
plt.plot([th, th], [-0.1, 1.1], color='k', linestyle='dashed', linewidth=2)
```

# Logistic model

Try to fit a model to the data. Assume each point is Bernoulli trial with probability $\pi_i$

First attempt: $\pi_i = x_i w$, problematic as LHS bounded between $[0,1]$

Second attempt: $odds_i = \pi_i / (1-\pi_i) = x_i w$, again bounded between $[0, \infty]$

Third attempt: $\log(odds_i) = log(\pi_i / (1-\pi_i)) = x_i w$ = great success! $[-\infty, \infty]$

Which gives rise to the logistic model: 

$$\pi_i = h_w(x_i) = \frac{1}{1 + e^{-x_i w}}$$

in our case:

$$h_w(x_i) = \frac{1}{1 + e^{-x_i w[1] - w[0]}}$$



```python
w = [0.0, -1.0]
```

```python

Y_PR = 1.0 / (np.exp(-w[1]* X - w[0]) + 1)
plt.scatter(X, Y_PR, color="red")
plt.scatter(X, Y, s=5)
plt.plot([th, th], [-0.1, 1.1], color='k', linestyle='dashed', linewidth=2)

```

# log-likelihood error function and its gradient

Assume that:

+ $P(y=1|x;w) = h_w(x)$
+ $P(y=0|x;w) = 1- h_w(x)$

or 

$$P(y|x;w) = h_w(x)^y(1-h_w(x))^{1-y}$$

Likelihood of all samples in data for a specific w (assuming data was generated IID):

$$L(w) = \prod(h_w(x_i)^{y_i}(1-h_w(x_i))^{1-y_i})$$

or

$$LL(w) = \sum y_i\log(h_w(x_i) + (1-y_i)\log(1-h_w(x_i))$$




```python
def pred(my_w):
    return 1.0 / (np.exp(-my_w[1]* X - my_w[0]) + 1)

def ll_loss(my_w):
    Y_PR = pred(my_w)
    ll = np.mean(-Y*np.log(Y_PR) - (1.0 - Y)*np.log(1.0-Y_PR))
    return ll
    
```

```python
SE,LL,GRAD = [],[],[]
w1 = -1.0
W0 = np.linspace(-20.0, 20.0, 50)
for w0 in W0:
    LL.append(ll_loss([w0, w1]))
    Y_PR = pred([w0, w1])
    GRAD.append(np.dot(Y_PR-Y, np.ones(len(X)))/len(Y_PR))
```

```python
plt.scatter(W0, GRAD, color="blue")
plt.plot([th, th], [-0.5, 0.5], color='k', linestyle='dashed', linewidth=2)
plt.show()    
plt.scatter(W0, LL, color="green")
for i in xrange(len(W0)):
    plt.arrow(W0[i], LL[i], -GRAD[i]*3, 0, head_width = 0.3)
plt.plot([th, th], [-0.5, 5.0], color='k', linestyle='dashed', linewidth=2)
plt.show()
```

# Gradient descent - batch version (all data is used to compute the gradient)

```python
W0 = [10.0]
L = [ll_loss([W0[0], w1])]
learning_rate = 1.0
EPS = 1e-6
it = 0
while it < 10 or (it < 1000 and abs(2*(L[-1]-L[-2])/(abs(L[-1])+abs(L[-2]))) > EPS):
    w0 = W0[-1]
    Y_PR = pred([w0, w1])
    grad = -np.dot(Y_PR-Y, np.ones(len(Y_PR)))/len(Y_PR)
    w0 = w0 + learning_rate * grad
    L.append(ll_loss([w0, w1]))    
    W0.append(w0)
    # put adaptive learning here...
    it += 1
print "iterations:", it, "w0 least square solution:", \
    W0[-1], "log loss:", ll_loss([W0[-1], w1])
plt.scatter(W0, L, c=range(len(W0)),  cmap='gray_r')
plt.plot([th, th], [-0.1, 1.0], color='k', linestyle='dashed', linewidth=2)
plt.show()
```

```python
my_s = random.sample(xrange(len(X)), 3)
plt.scatter(X, Y, s = 3)
plt.scatter(X[my_s], Y[my_s], color='red', s=30)
```

```python
def mini_pred(my_w, my_index):
    return 1.0 / (np.exp(-my_w[1]* X[my_index] - my_w[0]) + 1)

def mini_ll_loss(my_w, my_index):
    Y_PR = mini_pred(my_w, my_index)
    ll = np.mean(-Y[my_index]*np.log(Y_PR) - (1.0 - Y[my_index])*np.log(1.0-Y_PR))
    return ll
    
```

```python
W0 = [10.0]
learning_rate = 1.0
L = [ll_loss([W0[0], w1])]
it = 0
while it < 1000:
    w0 = W0[-1]
    my_s = random.sample(ind_train, 3)
    Y_PR = mini_pred([w0, w1], my_s)
    grad = -np.dot(Y_PR-Y[my_s], np.ones(len(Y_PR)))/len(Y_PR)
    delta = learning_rate*grad
    # put momentum here...
    w0 = w0 + delta
    ll = mini_ll_loss([W0[-1], w1], ind_validate)
    L.append(ll)
    W0.append(w0)
    
    if it % 10 == 9 and abs(np.mean(L)-np.mean(L[:-10])) < EPS:
        break

    if L[-1] < L[-2]:
        learning_rate *= 1.1
    else:
        learning_rate *= 0.5
        W0[-1] = W0[-2]    

    it += 1
    
print "iterations:", it, "w0 least square solution:", W0[-1], \
        "log loss:", ll_loss([W0[-1], w1])
plt.scatter(W0, L, c=range(len(W0)),  cmap='gray_r')
plt.plot([th, th], [-0.1, 2.0], color='k', linestyle='dashed', linewidth=2)
plt.show()
```

```python
def eval_numerical_gradient(f, x):
  fx = f(x) # evaluate function value at original point
  grad = np.zeros(x.shape)
  h = 0.00001
  # iterate over all indexes in x
  it = np.nditer(x, flags=['multi_index'], op_flags=['readwrite'])
  while not it.finished:
    # evaluate function at x+h
    ix = it.multi_index
    old_value = x[ix]
    x[ix] = old_value + h # increment by h
    fxh = f(x) # evalute f(x + h)
    x[ix] = old_value # restore to previous value (very important!)
    # compute the partial derivative
    grad[ix] = (fxh - fx) / h # the slope
    it.iternext() # step to next dimension

  return grad
```
