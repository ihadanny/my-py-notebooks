# happy lizzards

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline

img = mpimg.imread('lizzards.jpg')     
gray = np.dot(img, [0.299, 0.587, 0.144])   
def show_with_hist(img):    
    plt.imshow(img, cmap = plt.get_cmap('gray'))
    plt.show()
    plt.hist(img)
    plt.show()
show_with_hist(gray)
```

# oy vey! gaussian noise!

```python
noise = np.random.normal(scale=70.0, size=gray.shape)
noisy = gray + noise
show_with_hist(noisy)
```

# happy again with median filter

```python
import scipy.ndimage as ndimage
footprint = np.array([[1,1,1],
                      [1,0,1],
                      [1,1,1]])
denoise = ndimage.generic_filter(noisy, np.median, footprint=footprint)
show_with_hist(denoise)
```
