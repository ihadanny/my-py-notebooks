```python
name = 'B-large-practice'
f = open(name + '.in', 'r')
fo = open(name + '.out', 'w')

def remove_satisfied(l, milk, brand):
    new_l = []
    for cust in l:
        if milk in cust[brand]:
            pass
        else:
            for other_brand in ['0', '1']:
                if milk in cust[other_brand]:
                    cust[other_brand].remove(milk)
                new_l.append(cust)
    print len(new_l),
    return new_l
            
def shake_it(l):
    print len(l),
    if not l:
        "success! no more customers to satisfy"
        return {}
    l.sort(key=lambda x: len(x['0'])+ len(x['1']))
    problem_cust = l[0]
    for brand in ['0', '1']:
        for milk in problem_cust[brand]:
            new_l = remove_satisfied(l, milk, brand)
            son = shake_it(new_l)
            if son != "IMPOSSIBLE":
                son[milk] = brand
                return son
    return "IMPOSSIBLE"

C = int(f.readline())
for case in range(1, C+1):
    N = int(f.readline())
    M = int(f.readline())
    l = []
    for cust in range(M):
        cust_line = f.readline().split()
        cust_milks = {'0':set([]), '1':set([])}
        for milk in range(int(cust_line[0])):
            cust_milks[cust_line[2+milk*2]].add(cust_line[1+milk*2]) 
        l.append(cust_milks)
    print "start"
    answer = shake_it(l)
    if answer != "IMPOSSIBLE":
        str_answer = ""
        for milk in range(1, N+1):
            if str(milk) in answer:
                str_answer += answer[str(milk)] + " "
            else:
                str_answer += "0 "
    else:
        str_answer = answer
    msg = "Case #" + str(case) + ": " + str_answer
    print msg
    fo.write(msg + '\n') 
fo.close()
```

```python
import math
name = 'B-large-practice'
f = open(name + '.in', 'r')
fo = open(name + '.out', 'w')
C = int(f.readline())
for case in range(1, C+1):
    N, K, B, T = [ int(x) for x in f.readline().split() ] 
    X = [ int(x) for x in f.readline().split() ]
    V = [ int(x) for x in f.readline().split() ]
    Z = sorted(zip(X, V), reverse=True)
    price, in_home, blockers = 0, 0, 0
    for chick in Z:
        if chick[0] + chick[1]*T >= B:
            in_home += 1
            price += blockers
        else:
            blockers += 1
        if in_home >= K:
            break
    if in_home < K:
        price = "IMPOSSIBLE"
    msg = "Case #" + str(case) + ": " + str(price)
    fo.write(msg + '\n') 
fo.close()
```
