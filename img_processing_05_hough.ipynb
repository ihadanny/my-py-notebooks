# the homer

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import matplotlib.image as mpimg
%matplotlib inline

im = mpimg.imread("homer.jpg")
gray = np.dot(im, [0.299, 0.587, 0.144])  
plt.imshow(gray, cmap = plt.get_cmap('gray'))
```

# laplacian

```python
import scipy.ndimage as ndimage
def laplacian(x):
    return min(max(x[1] + x[3] - 4*x[4] + x[5] + x[7], 0),255)
edges = ndimage.generic_filter(gray, laplacian, size=(3,3))
def show_with_hist(img):    
    plt.imshow(img, cmap = plt.get_cmap('gray'))
    plt.show()
    plt.hist(img)
    plt.show()
show_with_hist(edges)
```

```python
import math
import operator
from collections import defaultdict
x_max, y_max = edges.shape[1], edges.shape[0]
votes = defaultdict(int)

print "hello loop"
increment = 4
c = 0
for index, val in np.ndenumerate(edges):
    if val > 250:
        for x0 in xrange(0, x_max, increment):
            for y0 in xrange(0, y_max, increment):
                rsqr = (index[1] - x0)**2 + (index[0] - y0)**2
                c += 1
                votes[(x0, y0, rsqr)] += 1
print "bye loop"
print c
sorted_votes = sorted(votes.items(), key=operator.itemgetter(1), reverse=True)
print "finished heavy lifting"
```

```python
fig = plt.figure()
fig = plt.gcf()
best_votes = [x[0] for x in sorted_votes[:5]] 
for v in best_votes:
    circ = plt.Circle((v[0], v[1]), radius=math.sqrt(v[2]), color='g', fill=False)
    fig.gca().add_artist(circ)
plt.imshow(gray, cmap = plt.get_cmap('gray'))
plt.show()
```
