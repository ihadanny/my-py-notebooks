# the homer

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import matplotlib.image as mpimg
%matplotlib inline

im = mpimg.imread("homer.jpg")
gray = np.dot(im, [0.299, 0.587, 0.144])  
plt.imshow(gray, cmap = plt.get_cmap('gray'))
```

# laplacian

```python
import scipy.ndimage as ndimage
def laplacian(x):
    return min(max(x[1] + x[3] - 4*x[4] + x[5] + x[7], 0),255)
edges = ndimage.generic_filter(gray, laplacian, size=(3,3))
def show_with_hist(img):    
    plt.imshow(img, cmap = plt.get_cmap('gray'))
    plt.show()
    plt.hist(img)
    plt.show()
show_with_hist(edges)
```

```python
import math
from collections import defaultdict
x_max, y_max = edges.shape[1], edges.shape[0]
r_max = max(x_max, y_max)
votes = defaultdict(int)
print "hello"
for index, x in np.ndenumerate(edges):
    if x > 250:
        for x0 in xrange(0, x_max):
            for y0 in xrange(0, y_max):
                r = int(math.sqrt((index[1] - x0)**2 + (index[0] - y0)**2))
                if r < r_max:
                    votes[(x0, y0, r)] += 1
print "bye"
import operator
sorted_votes = sorted(votes.items(), key=operator.itemgetter(1), reverse=True)
print sorted_votes[:5]
```
