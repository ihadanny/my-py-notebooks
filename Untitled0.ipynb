```python
a = [4, 10, 15, 24, 26]
b = [0, 9, 12, 20]
c = [5, 18, 22, 30]

def add_name(l, name):
    return [(x, name) for x in l]

def merge(a, b):
    if not a:
        return b
    if not b:
        return a
    if a[0] < b[0]:
        return [a[0]] + merge(a[1:], b)
    else:
        return [b[0]] + merge(a, b[1:])
        
abc = merge(merge(add_name(a, "a"), add_name(b, "b")), add_name(c, "c"))
print abc

def find_best_len(abc):
    best_start, best_end, best_len = 0, 0, 10000
    start, end = 0, 0
    while start < len(abc):
        dis_vals = set([x[1] for x in abc[start:end+1]])
        if len(dis_vals) == 3:
            if abc[end][0] - abc[start][0] < best_len:
                best_start, best_end = abc[start][0], abc[end][0]
                best_len = best_end - best_start
            if end <= start:
                return (best_start, best_end, best_len)
            else:
                start += 1
        else:
            if end < len(abc) - 1:
                end += 1
            else:
                start += 1
    return (best_start, best_end, best_len)

print find_best_len(abc)        
```

```python
import operator, sys

l = [17, 5, 40, 600, 3, 10, 54, 30, 100]

def play(l, money, alpha, beta, me):
    if len(l) == 1:
        if me:
            return money + l[0]
        else:
            return money
    if me:
        v = [0, 0]
        v[0] = play (l[1:], money + l[0], alpha, beta, not me)
        alpha = max(alpha, v[0])
        if alpha <= beta:
            v[1] = play(l[0:-1], money + l[-1], alpha, beta, not me)
        best_i, best = max(enumerate(v), key=operator.itemgetter(1)) 
    else:
        v = [sys.maxint, sys.maxint]
        v[0] = play (l[1:], money, alpha, beta, not me)
        beta = min(beta, v[0])
        if alpha <= beta:
            v[1] = play(l[0:-1], money, alpha, beta, not me)
        best_i, best = min(enumerate(v), key=operator.itemgetter(1))
    return best

print play(l, 0, -sys.maxint, sys.maxint, True)
```
