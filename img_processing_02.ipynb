# read the image into 8x8 tiles

```python
from PIL import Image
import numpy as np
import matplotlib
from matplotlib.pyplot import imshow

%matplotlib inline
im = Image.open("150110-173028.jpg")
imshow(np.asarray(im))

pixels = list(im.getdata())
width, height = im.size
eight = 8
tiles = {}

for y in xrange(height):
    for x in xrange(width):
        current_tile = (x/eight, y/eight)
        if not current_tile in tiles:
            tiles[current_tile] = np.zeros((eight,eight))
        tiles[current_tile][(y%eight)][x%eight] = pixels[y*width + x][0]            
```

# dct -> quantize -> idct

```python
from scipy.fftpack import dct, idct
import heapq

for key, tile in tiles.iteritems():
    d = dct(dct(tile,axis=0), axis=1)
    for y in xrange(eight):
        for x in xrange(eight):
            if x > 1 or y > 1: d[y][x] = 0
    tiles[key] = idct(idct(d, axis=1), axis=0)/256
```

# convert back to image

```python
pixels_wow = []        
for y in xrange(height):
    for x in xrange(width):
        current_tile = (x/eight, y/eight)
        p = (int(tiles[current_tile][y%eight][x%eight]),0,0)
        if x%eight == 1 or y%eight == 1: 
            p = (0,0,0)
        pixels_wow.append(p)


ima = Image.new(im.mode, im.size)
ima.putdata(pixels_wow)
imshow(np.asarray(ima))
```
